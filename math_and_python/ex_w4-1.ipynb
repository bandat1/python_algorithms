```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
from scipy.stats import maxwell
%matplotlib inline
```

# Распределение Парето
#### Cсылка на распределение: https://ru.wikipedia.org/wiki/%D0%A0%D0%B0%D1%81%D0%BF%D1%80%D0%B5%D0%B4%D0%B5%D0%BB%D0%B5%D0%BD%D0%B8%D0%B5_%D0%9F%D0%B0%D1%80%D0%B5%D1%82%D0%BE

#### Часть 1. Ниже сгенерирована выборка объёма 1000, построена гистограмма выборки и нарисована поверх неё теоретическая плотность распределения случайной величины.

```python
# выбраны параметры распределения
Xm = 1.0
b = 7.0

pareto_rv = sts.pareto(b)
sample = pareto_rv.rvs(1000)
```

```python
plt.hist(sample, normed=True, alpha=0.9)
x = np.linspace(1, 2, 100)
pdf = pareto_rv.pdf(x)
plt.plot(x, pdf, label='theoretical pdf', alpha=0.8)
plt.title('histogram of Pareto distribution')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
```

#### Часть 2.1. Ниже генерируются 1000 выборок объема n = 5. Расчитываются теоретические значения мат. ожидания и среднеквадратичного отклонения, для которых строится плотность нормального распределения. 

```python
# далее проводится расчет выборки для n=5
n = 5
matrix = []
for i in range (1000):
    sample = pareto_rv.rvs(n)
    matrix.append(sample.mean())

# далее проводится расчет среднего значения и среднеквадратичного отклонения по формулам из википедии
mu = (b * Xm) / (b - 1.0)
sigma = ((((Xm / (b - 1.0))**2) * b / (b - 2.0)) / n)**0.5

# задает нормальное распределение
norm_rv = sts.norm(loc=mu, scale=sigma) 

# построение диаграммы распределения Парето и теоретической плотности вероятности по нормальному расперелению 
plt.hist(matrix, normed=True, alpha=0.9)
x = np.linspace(1, 1.6, 100)
pdf = norm_rv.pdf(x)
plt.plot(x, pdf, label='norm distribution', alpha=0.8)
plt.title('histogram of chosen variables, n = 5')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```
